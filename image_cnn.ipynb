```
from os import listdir
import os
import pickle
from pickle import load
from pickle import dump
import numpy as np
import cv2

from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.models import Model
from keras.applications.vgg16 import decode_predictions

from google.colab.patches import cv2_imshow

import warnings
warnings.filterwarnings('ignore')
```

```
class cnnModel:
  def __init__(self):
    self.dirPath = "/content/drive/My Drive/NLP_Project/ProjectCode"
    self.dataPath = "/content/drive/My Drive/NLP_Project/ProjectCode/dataset"
    self.imgPath = "/content/drive/My Drive/NLP_Project/ProjectCode/dataset/Flicker8k_Dataset"
    self.modelsPath = "/content/drive/My Drive/NLP_Project/ProjectCode/models"
    self.trainDataFile = "Flickr_8k.trainImages.txt"
    self.testDataFile = "Flickr_8k.testImages.txt"
    
    self.model = VGG16()
    
  def getTrainImageNames(self):
    file = open(self.dataPath+"/"+self.trainDataFile, 'r')
    self.trainImgNames = []
    for line in file.read().split("\n"):
      self.trainImgNames.append(line)
    
    file.close()
    
  def prepareImage(self):
    self.model.layers.pop()
    model = Model(inputs=self.model.inputs, outputs=self.model.layers[-1].output)
  
    imageDict = {}

    i = 0
    
    self.getTrainImageNames()
    
    for imgName in self.trainImgNames:
      try:
        img = cv2.imread(os.path.join(self.imgPath,imgName))
        if(img is not None):
          img = cv2.resize(img,(224,224))
          img = img.reshape((1, img.shape[0], img.shape[1], img.shape[2]))
          imageDict[imgName] = self.model.predict(img)
      except Exception as e:
        print(str(e))

      i+=1
      
    dump(imageDict, open(self.modelsPath+'/preparedImages.pkl', 'wb'))
    print("done!")
    
  def extractFeatures(self):
    featuresDict = {}
    i = 0
    
    self.getTrainImageNames()
    
    for imgName in self.trainImgNames:
      try:
        img = cv2.imread(os.path.join(self.imgPath,imgName))
        if(img is not None):
          img = cv2.resize(img,(224,224))
          #cv2_imshow(img)
          img = img.reshape((1, img.shape[0], img.shape[1], img.shape[2]))
          featuresDict[imgName] = decode_predictions(self.model.predict(img))
      except Exception as e:
          print(str(e))
        
      i+=1
      
    dump(featuresDict, open(self.modelsPath+'/features.pkl', 'wb'))
    print("done!")
  
```

```
cnn = cnnModel()
```

```
cnn.prepareImage()
```

```
cnn.extractFeatures()
```
